```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        
        # 0을 만들 수 있는 3개의 엘리먼트를 출력하는 문제
        # 반복문 3개를 쓰는 건 Time complexity가 높아 Timeout 가능성이 크다.
        # 반복문 하나 + Two Pointer -> 정렬을 먼저 수행해야함.
        # Pointer가 끝에서부터 오는 것은 전체를 Search 하는 것
        # 앞에서부터 조합해가면서 풀이해보자.
        
        nums.sort() 
        # print(nums)
        results =[]
        length = len(nums) #시간 단축을 위함
        # for idx, val in enumerate(nums): # range로 바꾸면 시간 단축 가능
        for idx in range(length):
            # left가 0에서 시작하는게 아닌 idx +1에서 시작
            left, right = idx+1, length-1
            
            # 중복값 Pass:
            if idx>0 and nums[idx] == nums[idx-1]:
                continue            
            
            while left<right:
                ttl = nums[left]+nums[idx]+nums[right]
                
                if ttl >0: # right를 내리고
                    right -=1
                elif ttl<0 : # left를 올리고
                    left +=1
                else : 
                    results.append((nums[left],nums[idx],nums[right]))
                    #  만족할 때까지 반복
                                  
                    while left< right and nums[left]==nums[left+1]:
                        left +=1
                        
                    while left<right  and nums[right] == nums[right-1]:
                        right -=1
                    left +=1
                    right -=1
                    

        return results
```
